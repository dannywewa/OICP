```python
image = np.arange(0, 255, 10, dtype=np.uint8)
print(image/255)
print(ski.util.img_as_float(image))
```

```python
image = ski.data.coins()
rescaled = ski.transform.rescale(image, 0.5)
plt.imshow(rescaled)
plt.colorbar()
```
